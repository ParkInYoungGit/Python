# 매개변수를 추가해 요청을 전송하는 방법
기상청 RSS : http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp   
URL 뒤에 매개변수를 입력하여 지역별로 일기 정보를 가져올 수 있음.   
전국 : 108, 서울/경기도 : 109, 강원도 :105, 제주특별자치도 : 184   
ex) 서울/경기도   
http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109

```python
import urllib.request
import urllib.parse

API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

# 매개변수를 URL Encoding
values = {'stnId':'109'}
params = urllib.parse.urlencode(values)

# 요청 전용 URL 생성
url = API + "?" + params
print("url =", url)

data = urllib.request.urlopen(url).read()
#print(data) # 글자를 확인할 수 없음
text = data.decode('utf-8')
print(text)
```

# BeautifulSoup로 Scraping 하기
- BeautifulSoup : 간단하게 HTMLrhk XML에서 정보 추출
- Beautiful 설치 : pip install beautifulsoup4

```python
# 기본 사용법
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <h1> Header </h1>
                <p> Line 1을 서술 </p>
            </body>
        </html>
"""
# HTML 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
h1 = soup.html.body.h1
p = soup.html.body.p
print(h1)
print(p)

# text만 출력
print("h1 =", h1.string)
print("p =", p.string)
```

```python
# id로 요소를 추출하기
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <h1 id = "title"> Header </h1>
                <p id = "body"> Line 1을 서술 </p>
            </body>
        </html>
"""
# HTML 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
title = soup.find(id = "title")
body = soup.find(id = 'body')
print(title)
print(body)

# text만 출력
print('title =', title.string)
print('body=', body.string)

print('title =', title.text) # string이 안먹을 경우
```

```python
# 여러개의 요소를 추출하기
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <ul>
                    <li><a href ="http://www.naver.com">naver</a></li>
                    <li><a href ="http://www.daum.net">daum</a></li>
                </ul>
            </body>
        </html>
"""
# HTML 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
links = soup.find_all('a')
print(links)

for link in links:
    href = link.attrs['href'] # a태그 안에 있는 href
    text = link.string
    print(text, "==>", href)
```

```python
# 기상청 자료 활용하기
from bs4 import BeautifulSoup
import urllib.request as req

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

# data 가져오기
res = req.urlopen(url)

# BeautifulSoup 으로 분석
soup = BeautifulSoup(res, 'html.parser')
#print(soup)

# 원하는 데이터 추출하기
title = soup.find('title').string
wf = soup.find('wf').string

print(title)
print(wf)
```

```python
list(wf)
wfList = list(wf)
wfData = []
for i in wfList:
    if (i == '<')|(i == 'b')|(i == 'r')|(i == '/')|(i == '>'):
        continue
    else:
        
            wfData.append(i)
data = ""
for i in wfData:
    if i == "○":
        data = data + "\n"
    data = data + i

print(data)
```

```python
# 데이터 정리하기
list_wfs = list(wf)
except_char = ['<','b','r','/','>']
result = ''

for lwf in list_wfs:
    if lwf not in except_char:
        result += lwf

print(title)
print("=" * 50)
print(result)
```

# CSS 선택자 사용하기
- soup.select_one() : CSS 선택자로 요소 하나를 추출
- soup.select() : CSS 선택자로 여러개를 리스트로 추출

```python
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <div id = "meigen">
                    <h1>위키 북스</h1>
                    <ul class = "items">
                        <li>유니티 게임 이펙트 입문서</li>
                        <li>스위프트로 시작하는 아이폰 앱 개발교과서</li>
                        <li>모던 웹사이트 디자인의 정석</li>
                    </ul>
                </div>
            </body>
        </html>
"""

# HTML 분석하기
soup = BeautifulSoup(html, 'html.parser')

# 필요한 부분 CSS로 추출하기
# 타이틀 부분 추출하기 (id중복가능, class중복불가능)
h1 = soup.select_one("div#meigen > h1").string   # id:#, class:., tag:tag명, 자손:>, 후손:' '
print("h1 :", h1)

# 목록부분 추출하기
liDatas = soup.select("div#megin > ul.items > li")

for data in liDatas:
    print("li :", data.string)
```

# 네이버 금융에서 환율 정보 추출하기

https://finance.naver.com/marketindex/

```python
# 미국 환율 가져오기
from bs4 import BeautifulSoup
import urllib.request as req

# HTML 가져오기 (다운로드 받아서 분석하는 것)
url = "https://finance.naver.com/marketindex/"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, 'html.parser')

# 데이터 추출하기
price = soup.select_one("div.head_info > span.value").string
print(price)
```

```python
# 미국, 일본, 유럽연합, 중국의 환율 구하기

usdPrice = soup.select_one("a.usd > div.head_info > span.value").string
jpyPrice = soup.select_one("a.jpy > div.head_info > span.value").string
euroPrice = soup.select_one("a.eur > div.head_info > span.value").string
cnyPrice = soup.select_one("a.cny > div.head_info > span.value").string
print("미국환율 :", usdPrice)
print("일본환율 :", jpyPrice)
print("EU환율 :", euroPrice)
print("중국환율 :", cnyPrice)
```

```python
# 데이터 추출하기
prices = soup.select("div.head_info > span.value")

print("미국 :", prices[0].string)
print("일본 :", prices[1].string)
print("유럽연합 :", prices[2].string)
print("중국 :", prices[3].string)
```

# Rotten Tomatoes
### http://www.rottentomatoes.com/top/bestofrt/?year=2020

```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

# HTML 가져오기 (다운로드 받아서 분석하는 것)
url = "https://www.rottentomatoes.com/top/bestofrt/?year=2020"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, 'html.parser')


#top_movies_main > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > a
soup.select("#top_movies_main > div > table > tr > td > a")[0].string
```

## 적용

```python
li_lists = soup.select("#top_movies_main > div > table > tr > td > a")
count = 0
titleLists = [] # 영화제목 Lists
countLists = [] # Count Lists

for li in li_lists :
    count+= 1
    print(count, li.string.strip())
    countLists.append(count)
    titleLists.append(li.string.strip())
```

```python
# Lists확인
print(countLists)
print(titleLists)
```

```python
# countLists만으로 dataframe 생성
df = pd.DataFrame(data=countLists, columns = ['index'])
df
```

```python
# 기존 dafaframe에 titleLists 추가
df['Titles'] = pd.Series(titleLists)
df
```

```python
# csv로 저장
df.to_csv("../Data/rottenTomatoesList_2019.csv", encoding='utf-8', index=False)
```
